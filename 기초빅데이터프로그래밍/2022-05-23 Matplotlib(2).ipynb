```python
import numpy as np
import matplotlib.pyplot as plt
X=np.linspace(-np.pi,np.pi,256)
S=np.sin(X)
plt.plot(X,S,label='Sine')
plt.annotate(r'$(0,0)$',xy=(0,0),xytext=(-50,50),xycoords='data',textcoords='offset points',fontsize=16
            ,arrowprops=dict(arrowstyle='->',linewidth=3,color='g'))
plt.show()
```

```python
# Annotation
```

```python
plt.plot(X,S,label='sine')
plt.scatter([0],[0],color='r',linewidth=10)
plt.annotate(r'$(0,0)$',xy=(0,0),xytext=(-50,50),xycoords='data',textcoords='offset points',fontsize=16
            ,arrowprops=dict(arrowstyle='->',linewidth=3,color='g'))
plt.show()
```

```python
plt.plot(X,S,label='sine')
plt.scatter([0],[0],color='r',linewidth=10)
plt.grid(True)
plt.legend(loc=0)
plt.annotate(r'$(0,0)$',xy=(0,0),xytext=(-50,50),xycoords='data',textcoords='offset points',fontsize=16
            ,arrowprops=dict(arrowstyle='->',linewidth=3,color='g'))
plt.show()
```

```python
plt.plot(X,S,label='sine')
plt.scatter([0],[0],color='r',linewidth=10) # 산점도 처럼 점 찍기 
plt.annotate(r'$(0,0)$',xy=(0,0),xycoords='data', xytext=(0.3,0.7),textcoords='axes fraction',fontsize=16
            ,arrowprops=dict(arrowstyle='->',linewidth=3,color='g'))
plt.show()
```

```python
# Axes와 subplot
```

```python
fig=plt.figure()
ax1=fig.add_subplot(2,1,1) # add_subplot(행, 열 , 번호)
ax2=fig.add_subplot(2,1,2)
x=range(0,10)
y=[v**2 for v in x]
ax1.plot(x,y)
ax2.bar(x,y)
plt.show()
```

```python
x=range(0,10)
y=[v**2 for v in x]
plt.subplot(2,1,1)
plt.plot(x,y)
ax1=plt.gca()
plt.subplot(2,1,2)
plt.bar(x,y)
ax2=plt.gca()
print(ax1,id(ax1))
print(ax2,id(ax2))
plt.show()
```

```python
x1=np.linspace(0.0,5.0)
x2=np.linspace(0.0,2.0)
y1=np.cos(2*np.pi*x1)*np.exp(-x1)
y2=np.cos(2*np.pi*x2)

ax1=plt.subplot(2,1,1)
plt.plot(x1,y1,'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')
print(ax1)

ax2=plt.subplot(2,1,2)
plt.plot(x2,y2,'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')
print(ax2)

plt.show()
```
