# 주가를 DataFrame으로 저장해보기

```python
import numpy as np
import pandas as pd

daeshin={'open':[11650,11100,11200,11100,11000],"high":[12100,11800,11200,11100,11150],'low':[11600,11050,10900,10950,10900],
         'close':[11900,11600,11000,11100,11050]}

daeshin_day=pd.DataFrame(daeshin)
print(daeshin_day)
```

```python
date=['16.02.29','16.02.26','16.02.25','16.02.24','16.02.23']
daeshin_day=pd.DataFrame(daeshin,columns=['open','high','low','close'],index=date)
print(daeshin_day)
```

```python
x1=np.random.normal(1,1,size=(100,1))
x2=np.random.normal(-2,4,size=(100,1))
x=np.concatenate((x1,x2),axis=1)
x
df3=pd.DataFrame(x,columns=['x1','x2'])
df3
```

```python
# scatter 함수: 모양과 색 입히기
```

```python
x=np.random.rand(10)
y=np.random.rand(10)
z=np.sqrt(x**2+y**2)
plt.subplot(321)
plt.scatter(x,y,s=80,c=z,marker='>')
plt.show()
```
