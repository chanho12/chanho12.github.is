# x y좌표 둘다 표시

```python
plt.plot([10,20,30,40],[1,4,9,16])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-1.6,1.6,10000) # x좌표 나눈거 
f=lambda x: (np.sqrt(np.cos(x))*np.cos(200*x)+np.sqrt(abs(x))-0.7)/pow((4-x*x),0.01)
plt.plot(x,list(map(f,x)))
plt.show()
```

# 틱 설정

```python
X=np.linspace(-np.pi,np.pi,256)
Y=np.cos(X)
plt.plot(X,Y)
plt.show()
```

```python
X=np.linspace(-np.pi, np.pi,256)
Y=np.cos(X)
plt.plot(X,Y)
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi])
plt.yticks([-1,0,1])
plt.show()
```

# 그리드 설정

```python
X=np.linspace(-np.pi, np.pi,256)
Y=np.cos(X)
plt.plot(X,Y)
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi])
plt.yticks([-1,0,1])
plt.grid(True)
plt.show()
```

# 여러 개의 선을 그리기

```python
t=np.arange(0.,5.,0.2)
plt.plot(t,t,'r--',t,t**2,'bs:',t,t**3,'g^-') # y=x, y=x^2, y=x^3 을 그린거임 <정의역, 치역, 마커>
plt.show()
```

# Plot 겹쳐 그리기 : hold(True) 명령

```python
plt.plot([1,4,9,16],c='b',lw=5,ls='--',marker='p',ms=12,mec='g',mew=3,mfc='r')
plt.plot([9,16,4,1],c='k',lw=3,ls=':',marker='s',ms=8,mec='m',mew=3,mfc='c')
plt.show()
```

# Legend() 명령

```python
X=np.linspace(-np.pi,np.pi,256)
C=np.cos(X)
S=np.sin(X)
plt.plot(X,C,label='cosine')
plt.plot(X,S,label='sine')
plt.legend(loc=10)
plt.show()
```

```python
X=np.linspace(-np.pi,np.pi,256)
C=np.cos(X)
S=np.sin(X)
plt.plot(X,C,label='cosine')
plt.plot(X,S,label='sine')
plt.legend(loc=0)
plt.show()
```

# x축,y축 라벨, 타이틀

```python
X=np.linspace(-np.pi,np.pi,256)
C=np.cos(X)
S=np.sin(X)
plt.plot(X,C,label='cosine')
plt.xlabel('time')
plt.ylabel('amplitude')
plt.title('Cosine plot')
plt.show()
```

```python
X=np.linspace(-np.pi,np.pi,256)
C=np.cos(X)
plt.plot(X,C,label='cosine')
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi])
plt.yticks([-1,0,1])
plt.legend(loc=0)
plt.xlabel('times')
plt.ylabel('amplitude')
plt.title('Cosine Plot')
plt.show()
```
